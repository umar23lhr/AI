```python
target_result = 18
number = 0

while True:
    result = 0
    for digit in str(567):
        result += int(digit)

    if result == target_result:
        break

    number += 1
    target_result = target_result - result + number

print(f"The single-digit number that makes 567 equal to 18 is {number}")
```
