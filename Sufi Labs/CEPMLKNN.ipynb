```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import os
from os import listdir
```

```python
#train
Xtrain=[]
Ytrain=[]
leftSign="/content/drive/MyDrive/Train/LeftSign"
RightSign="/content/drive/MyDrive/Train/RightSign"
Stop="/content/drive/MyDrive/Train/Stop"
for images in os.listdir(leftSign):
    path = os.path.join(leftSign, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(1)
for images in os.listdir(RightSign):
    path = os.path.join(RightSign, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(0)
for images in os.listdir(Stop):
    path = os.path.join(Stop, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(2)
Xtrain=np.array(Xtrain)
Ytrain=np.array(Ytrain)
```

```python
#test
Xtest=[]
Ytest=[]
leftSign="/content/drive/MyDrive/Test/LeftSign"
RightSign="/content/drive/MyDrive/Test/RightSign"
Stop="/content/drive/MyDrive/Test/Stop"
for images in os.listdir(leftSign):
    path = os.path.join(leftSign, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(1)
for images in os.listdir(RightSign):
    path = os.path.join(RightSign, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(0)
for images in os.listdir(Stop):
    path = os.path.join(Stop, images)
    img=image.load_img(path)
    img2=img.resize((100,100))
    array=np.array(img2)
    X=array.reshape(-1)
    Xtrain.append(X)
    Ytrain.append(2)
Xtest=np.array(Xtest)
Ytest=np.array(Ytest)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import metrics
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
#KNN
K=2
knn = KNeighborsClassifier(n_neighbors=K)
knn.fit(Xtrain,Ytrain)

y_pred = knn.predict(Xtest)

    # Calculate evaluation metrics
accuracy = accuracy_score(Ytest, y_pred)
precision = precision_score(Ytest, y_pred, average='weighted')
recall = recall_score(Ytest, y_pred, average='weighted')
f1 = f1_score(Ytest, y_pred, average='weighted')
cm = confusion_matrix(Ytest, y_pred)

print("KNN Model with k=", K)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Plotting Confusion Matrix
confusion_test_plot = ConfusionMatrixDisplay(confusion_matrix=cm)
confusion_test_plot.plot()
plt.show()
```

```python
#Decsission Tree Classifictaion
from sklearn.tree import DecisionTreeClassifier
```

```python
km=DecisionTreeClassifier()
km.fit(Xtrain,Ytrain)
pre=km.predict(Xtest)

accuracy = accuracy_score(Ytest, pre)
precision = precision_score(Ytest, pre, average='weighted')
recall = recall_score(Ytest, pre, average='weighted')
f1 = f1_score(Ytest, pre, average='weighted')
cm = confusion_matrix(Ytest, pre)

print("Using decision tree classification")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Plotting Confusion Matrix
confusion_test_plot = ConfusionMatrixDisplay(confusion_matrix=cm)
confusion_test_plot.plot()
plt.show()
```
