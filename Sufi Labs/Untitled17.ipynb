```python
#reg . 200509 A Sufyan Mumtaz
import numpy as np
import pandas as pd
data = pd.read_csv('/content/loan_data.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,confusion_matrix,ConfusionMatrixDisplay
```

```python
X = data.iloc[:, :-1]
Y= data.iloc[:, -1]
```

```python
model = GaussianNB()
model.fit(X, Y)

pre = model.predict(X)
```

```python
print(pre)
```

```python
precision=precision_score(Y,pre)
acurracy=accuracy_score(Y,pre)
recall=recall_score(Y,pre)
F1=f1_score(Y,pre)
Con=confusion_matrix(Y,pre)

print(precision)
print(acurracy)
print(recall)
print(F1)
print(Con)
```
