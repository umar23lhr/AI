```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.tree import DecisionTreeClassifier as DT
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('/content/data lab5.csv')
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(df.drop('y',axis=1),df['y'],test_size=0.2,random_state=60)
```

```python
model=KNN(n_neighbors=2)
model.fit(Xtrain,Ytrain)
```

```python
predict=model.predict(Xtest)
predict
```

```python
precision= precision_score(Ytest, predict)
accuracy= accuracy_score(Ytest, predict)
recall = recall_score(Ytest, predict)
F1 = f1_score(Ytest, predict)
confusion=confusion_matrix(Ytest, predict)

print("Precision:", precision)
print("Accuracy:", accuracy)
print("Recall:", recall)
print("F1 Score:", F1)
print("Confusion Matrix:",confusion)
```

```python
confusio_test_plot = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion, display_labels = [False, True])
confusio_test_plot .plot()
plt.show()
```
