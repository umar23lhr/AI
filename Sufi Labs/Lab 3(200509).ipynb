```python
# Define a function to check if a word starts with an uppercase letter
def startsWithUppercase(word):
    if len(word) > 0 and word[0].isupper():
        return True
    else:
        return False

# Ask the user to input a word
user_input = input("Enter a word: ")

# Call the function and print the result
result = startsWithUppercase(user_input)

if result:
    print("True")
else:
    print("False")

```

```python
def joinNames(first_name, last_name):
    full_name = first_name + " " + last_name
    return full_name

# Example usage of the function
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")

full_name = joinNames(first_name, last_name)
print("Full Name:", full_name)
```
