```python
#SUFYAN MUMTAZ
#REF NO. 200509
#BEME-F-20-A
import numpy as np
import pandas as pd
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,confusion_matrix,ConfusionMatrixDisplay
```

```python
from sklearn.svm import SVC
```

```python
X=np.array([[3,1],[3,-1],[6,1],[6,-1],[1,0],[0,1],[0,-1],[-1,0]])
Y=np.array(['+','+','+','+','-','-','-','-'])
```

```python
svm=SVC(kernel='linear')
svm.fit(X,Y)
```

```python
coef=svm.coef_
intercept=svm.intercept_
```

```python
print("hyperplane equation")
print(f"{coef[0][0]}*x1+{coef[0][1]}*x2+{intercept}=0")
```

```python
pre=svm.predict(X)
```

```python

accuracy = accuracy_score(Y, pre)
precision = precision_score(Y, pre, pos_label='+')
recall = recall_score(Y, pre, pos_label='+')
f1 = f1_score(Y, pre, pos_label='+')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
