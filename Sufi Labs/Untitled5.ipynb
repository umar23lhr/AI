```python
names = ["Bob", "Canny", "amend", "Bob"]

unique_names = []

for name in names:
    if names.count(name) == 1:
        unique_names.append(name)

print(unique_names)
```

```python
word_list = []

while True:
    user_input = input("Enter a word (or 'quit' to exit): ")

    if user_input.lower() == 'quit':
        break

    word_list.append(user_input)

print("Words entered:")
for word in word_list:
    print(word)
```

```python
def starts_with_uppercase(word):
    if len(word) > 0:
        return word[0].isupper()
    return False


user_word = input("Enter a word: ")

result = starts_with_uppercase(user_word)

if result:
    print("True")
else:
    print("False")
```
