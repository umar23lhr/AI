```python
data_file = data_file.fillna(0)
data_file.dropna(axis=0)
data_file['title'] = data_file['title'].astype(str)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

green = '#52BE80'
red = '#EC7063'

# Create the plot
ax = sns.countplot(x=data_file['title'], palette=[green, red])

# Reduce image size
fig = plt.gcf()
fig.set_size_inches(10, 4)

# Display the plot
plt.tight_layout()
plt.show()
```
