```python
def ext_lsr(l, v):
  new_list = []
  for x in l:
    if x < v:
      new_list.append(x)
  return new_list
print(ext_lsr([1, 2, 3, 4, 5], 2))
```

```python
class Student:
  def __init__(self, name, cgpa):
    self.name = name
    self.cgpa = cgpa

# Create two Student objects
student1 = Student("Ali", 3.3)
student2 = Student("Ahmad", 3.7)

# Display the data for each Student
print("Student 1:")
print("Name:", student1.name)
print("CGPA:", student1.cgpa)

print("\nStudent 2:")
print("Name:", student2.name)
print("CGPA:", student2.cgpa)
```
