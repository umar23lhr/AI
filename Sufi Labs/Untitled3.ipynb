```python
def get_student_scores():

  scores = []
  for i in range(6):
    subject = input(f"Enter the score for subject {i + 1}: ")
    score = int(subject)
    scores.append(score)
    return scores


def calculate_average(scores):


  total = 0
  for score in scores:
    total += score
    average = total / len(scores)
    return average


def print_average_score(student_name, average_score):


  print(f"{student_name}'s average score is {average_score:.2f}.")

```
